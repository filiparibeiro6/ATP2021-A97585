```python
from random import randint
Lista=[]
maxi=[0]
def listamanual(Lista):
    num_elem=int(input("Introduza o número de elementos que deseja ter na lista:"))
    i=1
    while len(Lista) < num_elem :
        elemento=input("Introduza o elemento " +str(i) + ":")
        Lista.append(elemento)
        i=i+1
    return Lista
    
def listaaleatoria(Lista):
    num_elem=int(input("Introduza o número de elementos que deseja ter na lista:"))
    while len(Lista) < num_elem:
        elemento= randint (1,101)
        Lista.append(elemento)
    return Lista      
 
def bubbleSort(Lista): 
    for i in range(len(Lista)):
        trocas=0
        i=0
        while i<(len(Lista)-1):
            if int(Lista[i]) > int(Lista[i+1]):
                Lista[i],Lista[i+1] = Lista[i+1],Lista[i]
                trocas=1
            i=i+1
        if trocas== 0:
            return Lista
        
def maximo(Lista, maxi):
    for i in range(len(Lista)):
        i=0
        while(i <len(Lista)-1):
            if Lista[i] > Lista[i+1]:
                maxi.append(Lista[i])
                if Lista[i] > maxi[0]:
                    maxi.pop(0)
                else:
                    maxi.pop(1)
            else:
                maxi.append(Lista[i+1])
                if Lista[i+1] > maxi[0]:
                    maxi.pop(0)
                else:
                    maxi.pop(1)
            i=i+1
    return(maxi)

def soma1(Lista):
    j=0
    soma=0
    while (j < len(Lista)):
        n=int(Lista[j])
        soma=soma+n
        j=j+1
    return(soma)

def media(Lista):
    soma=soma1(Lista)
    media=soma/len(Lista)
    return(media)

def acrescentar(Lista):
    novo=int(input("Introduza o valor que deseja acrescentar:"))
    Lista.append(novo)
    return(Lista) 
    
def limparlista(Lista):
    Lista.clear()
    return Lista

"""----------------------------------//-----------------------------------"""
def menu(Lista, maxi):
    opcao=input("""Introduza uma das opções seguinte:
                
          (1)Gerar uma lista manualmente
          (2)Gerar uma lista de valores aleatórios
          (3)Ordenar a lista com bubbleSort
          (4)Calcular o valor máximo da lista
          (5)Calcular a soma de todos os elementos da lista
          (6)Calcular a média dos elementos da lista
          (7)Acrescentar à lista
          (8)Limpar a lista
          (0)Sair
          """)
    
    if opcao=="1":
        Lista= listamanual(Lista)
        print("A Lista é:", Lista)
        menu(Lista, maxi)
    elif opcao=="2":
        Lista=listaaleatoria(Lista)
        print("A Lista aleatória é:", Lista)
        menu(Lista, maxi)
    elif opcao=="3":
        Lista=bubbleSort(Lista)
        print("A Lista ordenada é:", Lista)
        menu(Lista, maxi)
    elif opcao=="4":
        maxi=maximo(Lista, maxi)
        print("O valor máximo da lista é:", maxi)  
        menu(Lista, maxi)
    elif opcao=="5":
        soma2= soma1(Lista)
        print("A soma dos elementos da lista é:", soma2)
        menu(Lista, maxi) 
    elif opcao=="6":
        media1=media(Lista)
        print("A média dos elementos da lista é:", media1)
        menu(Lista,maxi)
    elif opcao=="7":
        Lista1=acrescentar(Lista)
        print("A nova lista é:", Lista1)
        menu(Lista, maxi)
    elif opcao=="8":
        Lista=limparlista(Lista)
        print(Lista)
        menu(Lista, maxi)   
    elif opcao=="0":
        print("Até à próxima!")
    else:
        print("Nenhuma das opções!")
        menu(Lista, maxi)        
        
menu(Lista, maxi)    
```
