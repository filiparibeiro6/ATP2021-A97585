```python
def verfracao(Listafrac):
    numerador1,denominador1=Listafrac
    print("O resultado é:",str(numerador1)+"/"+str(denominador1))
```

```python
def simplificar(f):
    n,d=f
    m=mdc(n,d)
    return(int(n/m), int(d/m)) 
```

```python
def somar(f1,f2):
    n1, d1=f1
    n2, d2=f2
    n1=int(n1)
    d1=int(d1)
    n2=int(n2)
    d2=int(d2)
    return simplificar(((n1*d2)+(n2*d1),d1*d2))

def adicao(Listafrac):
    soma=Listafrac[0]
    for i in range (1,len(Listafrac)):
        soma=somar(soma, Listafrac[i])
    return soma
```

```python
def subtrair(f1,f2):
    n1, d1=f1
    n2, d2=f2
    n1=int(n1)
    d1=int(d1)
    n2=int(n2)
    d2=int(d2)
    return simplificar(((n1*d2)-(n2*d1),d1*d2))

def subtracao(Listafrac):
    subtracao=Listafrac[0]
    for i in range (1,len(Listafrac)):
        subtracao=subtrair(subtracao, Listafrac[i])
    return subtracao
```

```python
def dividir(f1,f2):
    n1, d1=f1
    n2, d2=f2
    n1=int(n1)
    d1=int(d1)
    n2=int(n2)
    d2=int(d2)
    if(d1==0 or d2==0):
        print("Impossivél!")
        menu(Listafrac)
    elif(n1==0 or n2==0):
        print("O resulatdo é:", 0)
        menu(Listafrac)
    
    else:
        return simplificar((n1*d2,n2*d1))

def divisao(Listafrac):
    divisao=Listafrac[0]
    for i in range (1,len(Listafrac)):
        divisao=dividir(divisao, Listafrac[i])
    return divisao
```

```python
def multiplicar(f1,f2):
    n1, d1=f1
    n2, d2=f2
    n1=int(n1)
    d1=int(d1)
    n2=int(n2)
    d2=int(d2)
    if(d1==0 or d2==0):
        print("Impossivél!")
        menu(Listafrac)
    elif(n1==0 or n2==0):
        print("O resulatdo é:", 0)
        menu(Listafrac)
    
    else:
        return simplificar((n1*n2,d1*d2))

def multiplicacao(Listafrac):
    multiplicacao=Listafrac[0]
    for i in range (1,len(Listafrac)):
        multiplicacao=multiplicar(multiplicacao, Listafrac[i])
    return multiplicacao
```

```python
def mdc(a,b):
    if b == 0:
        return a
    resto = a % b
    a = b
    b = resto
    return mdc(a, b)
```
